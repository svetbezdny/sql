### Динамика продаж

##### Дан датафрейм с продажами

```python
import pandas as pd
```

```python
df = pd.DataFrame({'sales': [100, 150, 150, 120, 135, 180, 230, 500, 501, 499, 500, 500]})
df
```

##### Задание:
К df добавить 2 новых столбца:
 - prev_sales - продажи за предыдущий период
 - inc - рост продаж по сравнению с предыдущим периодом
 
 Значения в столбце inc должны быть выражены в % и округлены до 2 знаков после запятой

**1. Добавляем столбец prev_sales со смещенными значениями на позицию вниз**

```python
df['prev_sales'] = df.shift(1)
df
```

**2. Добавляем столбец inc. Рассчитываем рост текущих продаж к предыдущим и округляем до сотых**

```python
df['inc'] = ((df['sales'] / df['prev_sales'] - 1) * 100).apply(lambda x: round(x, 2))
```
